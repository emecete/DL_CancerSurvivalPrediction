```r
brca_ex <- read.csv(file = 'brca_ex.csv', header = FALSE)
```

```r
genes <- read.csv(file = 'genes.txt', header = FALSE)
```

```r
genes <- genes[-length(genes)]
```

```r
as.vector(genes)
```

```r
library('caret')
```

```r
df2 = cor(brca_ex)
hc = findCorrelation(df2, cutoff=0.25) # putt any value as a "cutoff" 
hc = sort(hc)
reduced_Data = brca_ex[,-c(hc)]
```

```r
colnames(reduced_Data)
```

```r
genes[colnames(reduced_Data)]
```

```r
dim(reduced_Data)
```

```r
non_collinear_genes <- genes[colnames(reduced_Data)]
```

```r
write.csv(non_collinear_genes, 'non_collinear_genes.csv')
```
